```python
#importating liabraries
import pandas as pd
import numpy as np
```

```python
#loading data
df_ecommerce = pd.read_csv("data/messy_ecommerce_delivery_data.csv")
```

```python
#understaning the data
df_ecommerce.info()
df_ecommerce.head()
```

## From Above

```python
# checcking order id is unique
df_ecommerce[df_ecommerce["Order ID"].duplicated(keep = False)]
```

* There is 2100 duplicate order id 

```python
#removing duplicated id 
df_ecommerce = df_ecommerce.drop_duplicates( subset = "Order ID", keep = False)
```

```python
# rechecking the duplicates in order id
df_ecommerce [df_ecommerce["Order ID"].duplicated(keep = False)]
```

```python
#checking nulls in customer id
df_ecommerce[df_ecommerce["Customer ID"].isnull()]
```

```python
# checking unique values in platform
df_ecommerce["Platform"].unique()
```

```python
# converting all in lower case and removing extra spaces
df_ecommerce["Platform"] = df_ecommerce["Platform"].str.strip().str.lower()
```

```python
df_ecommerce["Platform"].unique()
```

```python
# changing the incorrect spelling
df_ecommerce["Platform"] = df_ecommerce["Platform"].replace({
    "jio mart" : "jiomart",
    "blinkt" : "blinkit"
})
```

```python
# rechecking
df_ecommerce["Platform"].unique()
```

```python
# filling nan value as n/a
df_ecommerce["Platform"] = df_ecommerce["Platform"].fillna("n/a")
```

```python
df_ecommerce["Platform"].isnull().sum()
```

```python
#making first letter capital
df_ecommerce["Platform"] = df_ecommerce["Platform"].str.capitalize()
```

```python
#understanding the date column
df_ecommerce["Order Date"].head()
```

```python
# removing the time from the order date
#converting order date to string
df_ecommerce["Order Date & Time"] = df_ecommerce["Order Date"].astype(str)

#spliting date and time 
df_ecommerce[["Order Date", "Order Time"]] = df_ecommerce["Order Date"].str.split(" ", n=1, expand=True)
```

```python
df_ecommerce.head()
```

```python
#changing date to date type
df_ecommerce["Order Date"] = pd.to_datetime(df_ecommerce["Order Date"], format = "mixed", errors = "coerce")
```

```python
#checking the type of date column
df_ecommerce["Order Date"].info()
```

```python
#Droping the order time and order date & time column
df_ecommerce.drop(columns = ["Order Time", "Order Date & Time"], inplace = True)
```

```python
df_ecommerce.head()
```

```python
#understanding delivery time column
df_ecommerce["Delivery Time (Minutes)"].info()
df_ecommerce["Delivery Time (Minutes)"].head()
```

```python
df_ecommerce["Delivery Time (Minutes)"].isnull().sum()
```

```python
# taking only number 
df_ecommerce["Delivery Time (Minutes)"] = df_ecommerce["Delivery Time (Minutes)"].astype(str).str.extract(r'(\d+\.?\d*)')[0]
```

```python
# converting into numeric
df_ecommerce["Delivery Time (Minutes)"] = pd.to_numeric(df_ecommerce["Delivery Time (Minutes)"], errors = "coerce")
```

```python
# filling null values usinng median
df_ecommerce["Delivery Time (Minutes)"].fillna(df_ecommerce["Delivery Time (Minutes)"].median(), inplace = True)
```

```python
df_ecommerce["Delivery Time (Minutes)"].describe()
```

```python
#checking the unique value in prdocut column
df_ecommerce["Product Category"].unique()
```

```python
#converting product category to lower case and removing extra space
df_ecommerce["Product Category"] = df_ecommerce["Product Category"].str.strip().str.lower()
```

```python
df_ecommerce["Product Category"].unique()
```

```python
# filling null value 
df_ecommerce["Product Category"].fillna("n/a", inplace = True)
```

```python
#checking null value
df_ecommerce["Product Category"].isnull().sum()
```

```python
# making first lettter capital
df_ecommerce["Product Category"] = df_ecommerce["Product Category"].str.capitalize()
```

```python
# understanding order value column 
df_ecommerce["Order Value (INR)"].info()
df_ecommerce["Order Value (INR)"].head()
```

```python
#extracting the number in correct order
df_ecommerce["Order Value (INR)"] = df_ecommerce["Order Value (INR)"].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
```

```python
# detecting outlier
Q1 = df_ecommerce['Order Value (INR)'].quantile(0.25)
Q3 = df_ecommerce['Order Value (INR)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_ecommerce[
    (df_ecommerce['Order Value (INR)'] < lower_bound) |
    (df_ecommerce['Order Value (INR)'] > upper_bound)
]

outliers
```

```python
#fixing the outlier
median = df_ecommerce["Order Value (INR)"].median()

df_ecommerce.loc[
    df_ecommerce["Order Value (INR)"] > upper_bound,
    "Order Value (INR)"
] = median
```

```python
df_ecommerce["Order Value (INR)"].describe()
```

```python
df_ecommerce["Order Value (INR)"].head()
```

```python
#filling null value using median
df_ecommerce["Order Value (INR)"] = df_ecommerce["Order Value (INR)"].fillna(df_ecommerce["Order Value (INR)"].median())
```

```python
df_ecommerce["Order Value (INR)"].isnull().sum()
```
